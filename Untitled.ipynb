```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv() 

client = OpenAI()
```

```python
response = client.images.generate(
  model="dall-e-3",
  prompt="a white siamese cat in a top hat",
  size="480x480",
  quality="standard",
  n=1,
)

image_url = response.data[0].url
```
